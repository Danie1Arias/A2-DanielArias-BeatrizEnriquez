```python
import matplotlib.pyplot as plt
from GeneticAlgorithm import GeneticAlgorithm 

data = [
    [4, 88, 8, 68, 6, 94, 5, 99, 1, 67],
    [2, 89, 9, 77, 7, 99, 0, 86, 3, 92],
    [5, 72, 3, 50, 6, 69, 4, 75, 2, 94],
    [8, 66, 0, 92, 1, 82, 7, 94, 9, 63],
    [9, 83, 8, 61, 0, 83, 1, 65, 6, 64],
    [5, 85, 7, 78, 4, 85, 2, 55, 3, 77],
    [7, 94, 2, 68, 1, 61, 4, 99, 3, 54],
    [6, 75, 5, 66, 0, 76, 9, 63, 8, 67],
    [3, 69, 4, 88, 9, 82, 8, 95, 0, 99],
    [2, 67, 6, 95, 5, 68, 7, 67, 1, 86]
]

# Hyperparameters
population_size = 150
generations = 200
mutation_rate = 0.3
elitism = 0.1

# Initialize the Genetic Algorithm
ga = GeneticAlgorithm(data, population_size=population_size, generations=generations, mutation_rate=mutation_rate, elitism=elitism)

# Run the algorithm
best_solution, best_makespan, evolution, decoded_solution = ga.run()
print("Best Solution (decoded):", decoded_solution)
print("Best Makespan:", best_makespan)

# Plot the evolution of the makespan
plt.plot(evolution, label="Best Makespan")
plt.xlabel("Generation")
plt.ylabel("Makespan")
plt.title("Evolution of Makespan Over Generations")
plt.legend()
plt.show()
```

```python
jobs = [
    {'machine': 0, 'start': 0, 'end': 2, 'name': 'job(0, 0)'},
    {'machine': 0, 'start': 2, 'end': 5, 'name': 'job(1, 0)'},
    {'machine': 1, 'start': 0, 'end': 3, 'name': 'job(0, 1)'},
    {'machine': 1, 'start': 3, 'end': 7, 'name': 'job(1, 1)'},
    {'machine': 1, 'start': 7, 'end': 9, 'name': 'job(2, 1)'},
    {'machine': 2, 'start': 0, 'end': 2, 'name': 'job(0, 2)'},
    {'machine': 2, 'start': 2, 'end': 6, 'name': 'job(1, 2)'},
    {'machine': 2, 'start': 9, 'end': 12, 'name': 'job(2, 2)'}
]
# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 3))

# Iterate through jobs and plot rectangles
for job in jobs:
    machine = job['machine']
    start = job['start']
    end = job['end']
    name = job['name']
    ax.barh(machine, end - start, left=start, height=0.5, label=name, alpha=0.7)
    ax.text(start + (end - start) / 2, machine, name, ha='center', va='center')

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Machine')
ax.set_title('Possible Solution of this Problem')

# Set y-axis limits
ax.set_yticks(range(len(jobs)))
ax.set_yticklabels([f'Machine {i}' for i in range(len(jobs))])

# Show the plot
plt.grid(axis='x')
plt.tight_layout()
plt.show()

```
